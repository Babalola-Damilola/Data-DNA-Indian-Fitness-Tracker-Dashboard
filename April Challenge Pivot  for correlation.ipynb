```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Importing the dataset
c = pd.read_excel("smartwatches.xlsx")
c
```

```python
# Checking datatypes
c.dtypes
```

```python
# convert 'Yes' to 1 and 'No' to 0 for the 'Bluetooth' and 'Touchscreen' columns
c['Bluetooth'] = c['Bluetooth'].replace({'Yes': 1, 'No': 0})
c['Touchscreen'] = c['Touchscreen'].replace({'Yes': 1, 'No': 0})
# convert Index from Int to object
c['Index'] = c['Index'].astype('object')
c.dtypes
```

```python
# Storing measure/numerical column names in "measure_cols" list
measure_cols = [i for i in c.select_dtypes(include=np.number).columns]
# Storing categorical column names in "cat_cols" list
cat_cols = [i for i in c.columns if i not in measure_cols]
measure_cols
```

```python
# Saving the pivoted
c.melt(id_vars=cat_cols).to_excel('smartwatces_pivot.xlsx')
# For display of the pivoted table
c.melt(id_vars=cat_cols)
```
